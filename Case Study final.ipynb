# Calculating Deposit of money on the platform, Withdrawal of money from the platform and Number of games played

```python
withdrawal['Withdrawal_Amount'] = withdrawal['Amount'] * 0.005
deposit['Deposit_Amount'] = deposit['Amount'] * 0.01
gameplay['Total_games_played'] = gameplay['Games Played'] * 0.2
```

```python
gameplay.rename(columns={'User ID': 'User Id'}, inplace=True)
```

```python
df = pd.merge(withdrawal, deposit, on = ['User Id', 'Datetime'], how = 'outer', suffixes=('_withdrawal', '_deposit'))
```

```python
df1 = pd.merge(df, gameplay, on = ['User Id', 'Datetime'], how = 'outer')
```

```python
df1.fillna(0, inplace=True)
```

# Calculating how many more times did a player do deposit than withdrawal

```python
df1['max_difference'] = df1['Deposit_Amount'] - df1['Withdrawal_Amount']
df1['More_deposit'] = df1['max_difference'].apply(lambda x: max(x * 0.001, 0))
df1.drop('max_difference', axis=1, inplace=True)
df1
```

```python
df1['loyalty_point'] = df1['Deposit_Amount'] + df1['Withdrawal_Amount'] + df1['More_deposit'] + df1['Total_games_played']
```

```python
df1
```

```python
# Calculating overall loyalty points for each player in October
overall_loyalty_points = df1.groupby('User Id')['loyalty_point'].sum().reset_index()
overall_loyalty_points.rename(columns={'loyalty_point': 'Overall_Loyalty_Points'}, inplace=True)

# Ranking players based on overall loyalty points
overall_loyalty_points['Rank'] = overall_loyalty_points['Overall_Loyalty_Points'].rank(ascending=False, method='min')

# Sorting by rank
overall_loyalty_points.sort_values(by='Rank', inplace=True)
```

```python
overall_loyalty_points
```

```python
# Calculatingverage deposit amount
average_deposit_amount = df1['Deposit_Amount'].mean()
average_deposit_amount
```

```python
# Calculating average deposit amount per user
average_deposit_per_user = df1.groupby('User Id')['Deposit_Amount'].mean().mean()
average_deposit_per_user
```

```python
# Calculating average number of games played per user
average_games_played_per_user = df1.groupby('User Id')['Games Played'].mean().mean()
average_games_played_per_user
```

# Part B - How much bonus should be allocated to leaderboard players?

```python
overall_loyalty_points_sorted = overall_loyalty_points.sort_values(by='Overall_Loyalty_Points', ascending=False)
```

```python
top_50_players = overall_loyalty_points_sorted.head(50)
```

```python
total_loyalty_points_top_50 = top_50_players['Overall_Loyalty_Points'].sum()
```

```python
top_50_players.loc[:, 'Bonus_Percentage'] = (top_50_players['Overall_Loyalty_Points'] / total_loyalty_points_top_50) * 100
```

```python
bonus_pool_amount = 50000
top_50_players.loc['Bonus_Amount'] = (top_50_players['Bonus_Percentage'] / 100) * bonus_pool_amount

print(top_50_players[['User Id', 'Bonus_Amount']])
```
